```python
# sentence = "Albert einstein was born in Ulm, Germany."
sentence = "The Arab Nationalists aimed, in part, to remove British imperial influence in Iraq."
```

```python

words = word_tokenize(sentence)

# Remove stop words
stop_words = set(stopwords.words('english'))
words = [word for word in words if word.lower() not in stop_words]

# Perform part-of-speech tagging
tagged_words = pos_tag(words)

ner_entities = ne_chunk(tagged_words)

n1 = ""
prev = ""
vec = []
for word, pos in tagged_words:
    if prev == "":
        if pos.startswith('N') or pos.startswith('J'):
            prev = pos
            n1 += (word + " ")
    elif prev.startswith('N') or prev.startswith('J'):
        if pos.startswith('N') or pos.startswith('J'):
            n1 += (word + " ")
        else:
            vec.append(n1.strip())
            prev = ""
            n1 = ""
print(vec)
# (Arab Nationalists, part, British, imperial influence, Iraq)
```

```python
from itertools import combinations
from collections import defaultdict

# Given array
arr = ["arab nationalist", "british", "imperial influence", "part", "iraq"]
# arr=vec

# Generate all possible combinations of tuples
all_tuples = []
for r in range(1, len(arr)):
    for comb_a in combinations(arr, r):
        remaining = [item for item in arr if item not in comb_a]
        for r_b in range(1, len(arr)):
            for comb_b in combinations(remaining, r_b):
                if not set(comb_a) & set(comb_b):
                    all_tuples.append((comb_a, comb_b))

# Filter out duplicates and reversed duplicates
unique_tuples = set()
for tup in all_tuples:
    sorted_tup = tuple(sorted(tup))
    unique_tuples.add(sorted_tup)

# Count tuple sizes (considering equivalence of (4,1) and (1,4))
size_counts = defaultdict(int)
for tup in unique_tuples:
    size_a = len(tup[0])
    size_b = len(tup[1])
    size_tuple = tuple(sorted((size_a, size_b)))
    size_counts[size_tuple] += 1

# Print counts
for size, count in size_counts.items():
    print(f"Tuple Size: {size}, Count: {count}")

# Sort unique tuples by tuple size
sorted_tuples = sorted(unique_tuples, key=lambda x: (len(x[0]), len(x[1])))

# Print sorted tuples
print("All Unique Tuples (Sorted by Tuple Size):")
for idx, tup in enumerate(sorted_tuples, start=1):
    print(f"{idx}. {tup}")
```
